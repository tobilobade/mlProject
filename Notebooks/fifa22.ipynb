```python
import pandas as pd

file_path = 'FIFA22_official_data.csv'
df = pd.read_csv(file_path)

# Step 5: Handle missing values
# Example strategy:
# - Drop columns with excessive missing values
# - Impute missing values for remaining columns

# Display columns before dropping
print("Columns before dropping:")
print(df.columns)
print("Total number of columns before dropping:", len(df.columns))

# Drop columns with excessive missing values
columns_to_drop = ['Marking', 'Loaned From']
df = df.drop(columns=columns_to_drop)

# Display columns after dropping
print("\nColumns after dropping:")
print(df.columns)
print("Total number of columns after dropping:", len(df.columns))

# Identify missing values before imputation
missing_values_before = df.isnull().sum()
print("Missing Values Before Imputation:")
print(missing_values_before[missing_values_before > 0])

# Impute missing values
for column in df.columns:
    if df[column].dtype in ['int64', 'float64']:
        # Fill missing values with mean for numerical columns
        df[column] = df[column].fillna(df[column].mean())
    elif df[column].dtype == 'object':
        # Fill missing values with mode for categorical columns
        df[column] = df[column].fillna(df[column].mode()[0])

# Identify missing values after imputation
missing_values_after = df.isnull().sum()
print("\nMissing Values After Imputation:")
print(missing_values_after[missing_values_after > 0])
```
